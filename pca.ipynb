# Principal Component Analysis of Grimm Brothers

## Setup and Imports

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.preprocessing import normalize
from sklearn.decomposition import PCA, TruncatedSVD as SVD
```

```python
import plotly.express as px
import plotly.io as pio
```

```python
TFIDF = pd.read_csv('data/pg2591-TFIDF.csv',index_col=0)
TFIDF.head()
```

```python
DTM = pd.read_csv('data/pg2591-DTM.csv',index_col=0)
DTM.head()
```

```python
LIB = pd.read_csv('data/pg2591-LIB.csv',index_col=0)
LIB.head()
```

## DCM

```python
pca_engine = PCA(n_components=5)
DCM = pd.DataFrame(pca_engine.fit_transform(TFIDF), index=TFIDF.index)
DCM.columns = ['PC{}'.format(i) for i in DCM.columns]
DCM
```

```python
## Make Loading
```

```python
LOADING = pd.DataFrame(pca_engine.components_.T * np.sqrt(pca_engine.explained_variance_))
LOADING.columns = ["PC{}".format(i) for i in LOADING.columns]
LOADING.index = TFIDF.columns
LOADING.index.name = 'term_str'
LOADING.head()
```

## Get Top 5 Postive Terms for PC0

```python
LOADING.sort_values(by='PC0',ascending=False).head(5)
```

## Get Top 5 Negatuve Terms for PC1

```python
LOADING.sort_values(by='PC1',ascending=True).head(5)
```

## Creating Visuals

## Getting DOC Table ready

```python
DOC = DTM.sum(1).to_frame('n_tokens')
DOC['n_types'] = DTM.astype('bool').sum(1)
DOC['pkr'] = DOC.n_types / DOC.n_tokens
DOC = DOC.join(LIB['volume'])
```

```python
NEW=DCM.join(DOC)
NEW
```

### Importing Functions

```python
def vis_pcs(a, b,DCM, col='doc_label'):
    fig =px.scatter(DCM, 
        f"PC{a}", f"PC{b}", 
        color=DOC[col].astype('category'), 
        hover_name=DCM.index,
        marginal_x='box', 
        height=1000, 
        width=1200)
    fig.show()
```

```python
def vis_loadings(COMPS, a=0, b=1, hover_name='term_str'):
    return px.scatter(COMPS.reset_index(), f"PC{a}", f"PC{b}", 
                      text='term_str', 
                      marginal_x='box', 
                      height=800)
```
